```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


data = pd.read_csv('datasetvis.csv')

data['TIME_PERIOD'] = pd.to_numeric(data['TIME_PERIOD'], errors='coerce')
data['OBS_VALUE'] = pd.to_numeric(data['OBS_VALUE'], errors='coerce')

# filter fiji
fiji_data = data[data['GEO_PICT'] == 'FJ'].dropna(subset=['OBS_VALUE'])

plt.figure(figsize=(8,5))
sns.lineplot(x='TIME_PERIOD', y='OBS_VALUE', data=fiji_data, marker='o')
plt.title('Line Plot - OBS_VALUE di Fiji dari waktu ke waktu')
plt.xlabel('Tahun')
plt.ylabel('OBS_VALUE')
plt.grid(True)
plt.show()
```

Grafik di atas menunjukkan perkembangan nilai OBS_VALUE di negara Fiji dari waktu ke waktu, mulai dari tahun 1999 hingga 2023. Secara umum, terdapat tren peningkatan yang fluktuatif, di mana nilai OBS_VALUE mengalami naik-turun setiap tahunnya namun cenderung meningkat pada periode akhir pengamatan.

Terlihat bahwa pada sekitar tahun 2011, terjadi lonjakan signifikan dalam nilai OBS_VALUE, yang kemudian diikuti oleh penurunan tajam di tahun-tahun berikutnya. Setelah itu, nilai cenderung stabil meskipun tetap mengalami fluktuasi kecil. Lonjakan lain yang cukup mencolok terjadi pada tahun 2023, yang mencatatkan nilai OBS_VALUE tertinggi selama periode pengamatan.

Area biru yang mengelilingi garis menunjukkan variasi atau sebaran data, yang semakin melebar pada tahun-tahun tertentu. Hal ini menandakan bahwa pada periode tersebut terdapat ketidakstabilan atau variasi tinggi dalam nilai OBS_VALUE, khususnya di sekitar tahun 2010 dan 2023.
