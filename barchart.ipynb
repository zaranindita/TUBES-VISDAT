```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('datasetvis.csv')

data['OBS_VALUE'] = pd.to_numeric(data['OBS_VALUE'], errors='coerce')

# Hitung rata-rata
top_countries = data.groupby('GEO_PICT')['OBS_VALUE'].mean().sort_values(ascending=False).head(5)

# Bar Chart
plt.figure(figsize=(8,5))
sns.barplot(x=top_countries.index, y=top_countries.values, palette='viridis')
plt.title('Bar Chart - Rata-rata OBS_VALUE 5 Negara Teratas')
plt.xlabel('Kode Negara')
plt.ylabel('Rata-rata OBS_VALUE')
plt.show()
```
