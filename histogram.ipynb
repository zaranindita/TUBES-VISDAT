```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('datasetvis.csv')

data['OBS_VALUE'] = pd.to_numeric(data['OBS_VALUE'], errors='coerce')

# Histogram
plt.figure(figsize=(8,5))
sns.histplot(data['OBS_VALUE'].dropna(), bins=30, kde=True, color='skyblue')
plt.title('Histogram - Distribusi OBS_VALUE')
plt.xlabel('OBS_VALUE')
plt.ylabel('Frekuensi')
plt.show()
```
