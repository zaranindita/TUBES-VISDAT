```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('datasetvis.csv')

data['OBS_VALUE'] = pd.to_numeric(data['OBS_VALUE'], errors='coerce')

# top 5 rata2 obs value
top_countries = data.groupby('GEO_PICT')['OBS_VALUE'].mean().sort_values(ascending=False).head(5)
top_5_countries = top_countries.index.tolist()

top5_data = data[data['GEO_PICT'].isin(top_5_countries)]

plt.figure(figsize=(8,5))
sns.boxplot(x='GEO_PICT', y='OBS_VALUE', data=top5_data, palette='Set2')
plt.title('Boxplot - OBS_VALUE pada 5 Negara Teratas')
plt.xlabel('Kode Negara')
plt.ylabel('OBS_VALUE')
plt.show()
```

Gambar boxplot di atas menunjukkan distribusi nilai OBS_VALUE dari lima negara dengan rata-rata tertinggi, yaitu PG, VU, PF, SB, dan NC. Berdasarkan visualisasi, terlihat bahwa negara NC dan PF memiliki sebaran data yang sangat luas, ditandai dengan banyaknya titik outlier yang tersebar jauh dari nilai tengah. Hal ini mengindikasikan adanya nilai-nilai ekstrem yang sangat tinggi dan rendah dalam data mereka.

Sebaliknya, negara PG, VU, dan SB memiliki distribusi nilai OBS_VALUE yang lebih sempit dan terkonsentrasi, dengan jumlah outlier yang lebih sedikit. Ini menunjukkan bahwa data dari ketiga negara tersebut cenderung lebih stabil dan tidak terlalu bervariasi.

Dari keseluruhan visualisasi, dapat disimpulkan bahwa meskipun NC dan PF memiliki nilai rata-rata OBS_VALUE yang tinggi, data mereka juga mengandung variasi yang besar dan banyak outlier. Sementara negara lain menunjukkan pola distribusi yang lebih konsisten dan tidak terlalu ekstrem.
