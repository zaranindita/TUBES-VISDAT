```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('datasetvis.csv')

# jumlah data per negara
top_counts = data['GEO_PICT'].value_counts().head(5)

plt.figure(figsize=(6,6))
plt.pie(top_counts.values, labels=top_counts.index, autopct='%1.1f%%', colors=sns.color_palette('pastel'))
plt.title('Pie Chart - Distribusi Jumlah Data per Negara (Top 5)')
plt.show()
```

Diagram lingkaran menunjukkan distribusi jumlah data berdasarkan negara untuk lima negara teratas. Negara dengan kontribusi jumlah data terbesar adalah FJ (Fiji) dengan proporsi sebesar 28,4%, menunjukkan bahwa sebagian besar data berasal dari Fiji.

Negara TO (Tonga) menyumbang 19,9%, diikuti oleh KI (Kiribati) sebesar 19,0%, dan VU (Vanuatu) sebesar 17,9%. Sementara itu, negara dengan proporsi terkecil dari lima besar adalah TV (Tuvalu) dengan 14,8%.

Distribusi ini menunjukkan bahwa meskipun terdapat lima negara yang mendominasi, kontribusi mereka terhadap jumlah data relatif merata, dengan Fiji sebagai negara yang paling menonjol. Hal ini bisa menjadi indikasi bahwa Fiji memiliki sistem pelaporan atau pencatatan data yang lebih konsisten atau intensif dibandingkan negara-negara lain di kawasan tersebut.
